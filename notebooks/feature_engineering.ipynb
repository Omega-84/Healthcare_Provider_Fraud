```python
import numpy as np
import pandas as pd
from warnings import filterwarnings
filterwarnings("ignore")
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 20)
```

```python
train_inpatient = pd.read_csv("../data/Train_Inpatientdata-1542865627584.csv")
train_beneficiary = pd.read_csv("../data/Train_Beneficiarydata-1542865627584.csv")
train_outpatient = pd.read_csv("../data/Train_Outpatientdata-1542865627584.csv")
train_provider = pd.read_csv("../data/Train-1542865627584.csv")
```

## Beneficiary Data Processing

```python
train_beneficiary.loc[:, train_beneficiary.columns.str.contains('ChronicCond_')] = \
train_beneficiary.filter(like='ChronicCond_').replace({1: 1, 2: 0})
train_beneficiary = pd.concat([train_beneficiary,
 pd.get_dummies(train_beneficiary['Gender'].map({1: 'Male', 2: 'Female'}), prefix='Gender'),
 pd.get_dummies(train_beneficiary['Race'].map({1: 'White', 2: 'Black', 3: 'Other', 4: 'Unknown', 5: 'Hispanic'}), prefix='Race')], axis=1)
train_beneficiary['RenalDiseaseIndicator'] = train_beneficiary['RenalDiseaseIndicator'].map({'0':0,'Y':1})
```

## Feature Creation

```python
train_inpatient['Is_Inpatient'] = True
train_inpatient.head()
```

```python
train_outpatient['Is_Outpatient'] = True
train_outpatient.head()
```

```python
patient_data = pd.concat([train_inpatient, train_outpatient])
```

```python
patient_data
```

```python
for i in ['ClaimStartDt','ClaimEndDt','AdmissionDt','DischargeDt']:
    patient_data[i] = pd.to_datetime(patient_data[i])
```

```python
patient_data['billing_before_admission'] = (patient_data['ClaimStartDt'] - patient_data['AdmissionDt']).dt.days < 0            
patient_data['billing_after_discharge'] = (patient_data['ClaimEndDt'] - patient_data['DischargeDt']).dt.days > 0
```

```python
top_diagnosis_code = patient_data.groupby(['ClmAdmitDiagnosisCode'])['ClaimID'].count().reset_index().sort_values(by='ClaimID', ascending=False).head(20)['ClmAdmitDiagnosisCode'].unique()
```

```python
pd.DataFrame(top_diagnosis_code,columns=['ClmAdmitDiagnosisCode']).to_csv("../data/top_diagnosis_code.csv",index=False)
```

```python
patient_data['top_diagnosis_code'] = np.where(patient_data['ClmAdmitDiagnosisCode'].isin(top_diagnosis_code),1,0)
```

```python
patient_data['hospital_stay_days'] = (patient_data['DischargeDt'] - patient_data['AdmissionDt']).dt.days + 1
patient_data['num_diagnoses'] =patient_data.filter(like='ClmDiagnosisCode_').notna().sum(axis=1)
patient_data['num_procedures'] =patient_data.filter(like='ClmProcedureCode_').notna().sum(axis=1)
```

```python
patient_data['claim_duration_days'] = (patient_data['ClaimEndDt'] - patient_data['ClaimStartDt']).dt.days + 1
patient_data['num_of_physicians'] = patient_data.filter(like='Physician').notna().sum(axis=1)
# Top 20 for each role separately
top_20_attending = patient_data['AttendingPhysician'].value_counts().head(20).index
top_20_operating = patient_data['OperatingPhysician'].value_counts().head(20).index

# Binary columns
patient_data['is_top20_attending'] = patient_data['AttendingPhysician'].isin(top_20_attending).astype(int)
patient_data['is_top20_operating'] = patient_data['OperatingPhysician'].isin(top_20_operating).astype(int)
patient_data['stay_vs_claim_diff'] = patient_data['hospital_stay_days'] - patient_data['claim_duration_days']
```

```python
pd.DataFrame(top_20_attending,columns=['AttendingPhysician']).to_csv("../data/top_attending_physician.csv",index=False)
pd.DataFrame(top_20_operating,columns=['OperatingPhysician']).to_csv("../data/top_operating_physician.csv",index=False)
```

```python
patient_data = patient_data.merge(train_beneficiary,on='BeneID',how='left')
patient_data['DOB'] = pd.to_datetime(patient_data['DOB'])
patient_data['Age_at_Claim'] = ((pd.to_datetime(patient_data['ClaimStartDt']) - pd.to_datetime(patient_data['DOB'])).dt.days / 365.25).astype(int)
```

```python
patient_data.head()
```

```python
patient_data['Is_Dead'] = np.where(patient_data['DOD'].isna(), False, True)
```

```python
for i in ['hospital_stay_days','stay_vs_claim_diff']:
    patient_data[i] = patient_data[i].fillna(0)
for i in ['Is_Inpatient','Is_Outpatient']:
    patient_data[i] = patient_data[i].fillna(False)
```

```python
features = patient_data.groupby(['Provider']).agg(
    count_unique_beneficiary = ('BeneID','nunique'),
    count_unique_claims = ('ClaimID','count'),
    count_dead_beneficiary = ('Is_Dead','sum'),
    count_unique_states = ('State','nunique'),
    count_unique_counties = ('County','nunique'),
    mean_hospital_stay_days = ('hospital_stay_days', 'mean'),
    max_hospital_stay_days = ('hospital_stay_days', 'max'),
    total_top_diagnosis_codes = ('top_diagnosis_code','sum'),
    mean_total_diagnosis =('num_diagnoses','mean'),
    mean_total_procedures = ('num_procedures','mean'),
    mean_number_of_physicians = ('num_of_physicians','mean'),
    mean_difference_stay_vs_claim = ('stay_vs_claim_diff','mean'),
    patients_under_top_attending_physician = ('is_top20_attending','sum'),
    patients_under_top_operating_physician = ('is_top20_operating','sum'),
    mean_claim_amount = ('InscClaimAmtReimbursed','mean'),
    total_claim_amount = ('InscClaimAmtReimbursed','sum'),
    std_claim_amount = ('InscClaimAmtReimbursed','std'),
    mean_age = ('Age_at_Claim','mean'),
    count_of_males = ('Gender_Male','sum'),
    count_of_females = ('Gender_Female','sum'),
    count_of_black_people = ('Race_Black','sum'),
    count_of_white_people = ('Race_White','sum'),
    count_of_hispanic_people = ('Race_Hispanic','sum'),
    count_of_other_people = ('Race_Other','sum'),
    count_alzheimer = ('ChronicCond_Alzheimer', 'sum'),
    count_heartfailure = ('ChronicCond_Heartfailure', 'sum'),
    count_kidneydisease = ('ChronicCond_KidneyDisease', 'sum'),
    count_cancer = ('ChronicCond_Cancer', 'sum'),
    count_obstrpulmonary = ('ChronicCond_ObstrPulmonary', 'sum'),
    count_depression = ('ChronicCond_Depression', 'sum'),
    count_diabetes = ('ChronicCond_Diabetes', 'sum'),
    count_ischemicheart = ('ChronicCond_IschemicHeart', 'sum'),
    count_osteoporasis = ('ChronicCond_Osteoporasis', 'sum'),
    count_rheumatoidarthritis = ('ChronicCond_rheumatoidarthritis', 'sum'),
    count_stroke = ('ChronicCond_stroke', 'sum'), 
    count_renal_disease = ('RenalDiseaseIndicator', 'sum'),   
    mean_annual_inpatient_reimbursement = ('IPAnnualReimbursementAmt','mean'),
    mean_annual_inpatient_deductible = ('IPAnnualDeductibleAmt','mean'),
    mean_annual_outpatient_reimbursement = ('OPAnnualReimbursementAmt','mean'),
    mean_annual_outpatient_deductible = ('OPAnnualDeductibleAmt','mean'),
    count_inpatient = ('Is_Inpatient','sum'),
    count_outpatient = ('Is_Outpatient','sum')
).reset_index()
```

```python
features['claims_per_bene'] = features['count_unique_claims'] / features['count_unique_beneficiary']
```

```python
train_provider = train_provider.merge(features, on="Provider", how="left")
```

```python
train_provider.head()
```

```python
train_provider.isna().sum().reset_index().rename(columns={"index": "feature", 0: "missing_values"}).sort_values("missing_values", ascending=False)
```

```python
train_provider['std_claim_amount'] = train_provider['std_claim_amount'].fillna(0)
```

```python
train_provider['PotentialFraud'] = train_provider['PotentialFraud'].map({"No": 0, "Yes": 1})
```

```python
train_provider.groupby(['PotentialFraud']).agg(
    mean_count_unique_states=('count_unique_states','mean'),
    mean_count_unique_counties=('count_unique_counties','mean'),
    mean_count_unique_claims=('count_unique_claims','mean'),
    mean_count_unique_beneficiaries=('count_unique_beneficiary','mean'),
    mean_claim_amount=('mean_claim_amount','mean'),
    mean_beneficiary_age=('mean_age','mean')
)
```
